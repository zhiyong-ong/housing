```python
with open('data/MacAdd.txt', 'r') as f:
    data = f.read()
data_list = data.split('\n')
data_start_idx = 0
for idx, item in enumerate(data_list):
    if '---------' in item:
        data_start_idx = idx + 1
        break

mac_addr_list = []
for i in range(data_start_idx, len(data_list)):
    mac_addr = data_list[i].strip().split(' ')[0]
    if mac_addr:
        mac_addr_list.append(mac_addr)
```

```python
output = []
output.append('<?xml version="1.0" encoding="UTF-8" standalone="yes"?>')
output.append('<TipsContents xmlns="http://www.avendasys.com/tipsapiDefs/1.0">')
output.append('\t<TipsHeader exportTime="Thu Mar 05 11:17:42 SGT 2020" version="6.7"/>')
output.append('\t<StaticHostLists>')
output.append('\t\t<StaticHostList description="" name="A1L6ASW003-005" memberType="MACAddress" memberFormat="list">')
output.append('\t\t\t<Members>')
for mac in mac_addr_list:
    out_str = f'\t\t\t\t<Member description="Camera" address="{mac}"/>'
    output.append(out_str)
output.append('\t\t\t</Members>')
output.append('\t\t</StaticHostList>')
output.append('\t</StaticHostLists>')
output.append('</TipsContents>')

with open('data/output_file.xml', '+w') as f:
    f.write('\n'.join(output))
```
